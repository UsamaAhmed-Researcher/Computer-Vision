## Read Images Video Webcam

### Image

This Python code uses the OpenCV library to read and display an image named `"lenna.png"`. The cv2.imread function loads the image, and `cv2.imshow` displays it in a window titled `"Lena Soderberg"`. The `cv2.waitKey(0)` line waits for a key press before closing the window

```python
import matplotlib.pyplot as plt
img = cv2.imread("lenna.png")

plt.imshow(img)
#cv2.waitKey(0)
```

### Webcam

This code of OpenCV library to capture video from the default webcam (camera index 0). It sets the frame width and height using the `cap.set()` function. The captured frames are displayed in a window titled "Result" using `cv2.imshow`. Pressing the 'q' key exits the video capture loop.

```python
frameWidth = 640
frameHeight = 480

cap = cv2.VideoCapture(0)

cap.set(3, 640)
cap.set(4, 480)

while True:
    _, img = cap.read()
    cv2.imshow("Result", img)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

## Basic Functions


### Grey Scale

This line of code converts the color image `img` from the BGR color space to grayscale using OpenCV's cvtColor function. The resulting grayscale image is stored in the variable imgGray, which can be used for further processing or display

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

### Gaussian Blur

The GaussianBlur function in OpenCV is used for smoothing an image to reduce noise and details. The parameters (7, 7) specify the kernel size for blurring, and 0 indicates the standard deviation of the Gaussian kernel (in this case, it's automatically calculated based on the kernel size). The resulting blurred image is stored in the variable `imgBlur`.

```python
imgBlur = cv2.GaussianBlur(imgGray, (7, 7), 0)
plt.imshow(imgBlur)
```

### Image Canny

Canny edge detection algorithm to the input image img. The Canny function in OpenCV detects edges in the image by first applying Gaussian smoothing and then finding the intensity gradients. The parameters 150 and 200 represent the lower and upper thresholds for edge detection. Any gradient value above the upper threshold is considered an edge, and any value below the lower threshold is considered not an edge.

1. Lowering the low threshold will result in more edges being detected, including weaker edges or noise in the image.
2. Increasing the high threshold will result in fewer strong edges being detected as only edges with stronger gradients will be considered.

```python
imgCanny = cv2.Canny(imgBlur,150,300)
plt.imshow(imgCanny)
```

## Shapes and Text

### Line

It draws a green line `((0, 255, 0) represents the color in BGR format)` on the image img. The line starts from coordinates `(0, 0)` (top-left corner) and ends at coordinates `(200, 250)` (bottom-right corner). The line thickness is 3 pixels.

```python
img = cv2.imread("lenna.png")
newimg = cv2.line(img,(20,0),(200,250),(0,255,0),3)
plt.imshow(newimg)
```

### Rectangle

This code draws a red rectangle `((0, 0, 255) represents the color in BGR format)` on the image img. The rectangle is defined by its top-left corner `(0, 0)` and its bottom-right corner `(250, 350)`. The rectangle has a border thickness of 2 pixels.

```python
img = cv2.imread("lenna.png")
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
newimg = cv2.rectangle(imgrgb, (50, 50), (250, 350), (0, 0, 0), 5)
plt.imshow(newimg)
```

### Circle

This line draws a circle on the image img. The circle's center is at `(400, 50)`, its radius is 30 pixels, and its color is yellow `((255, 255, 0) in BGR format)`. The circle's border has a thickness of 5 pixels.

```python
cv2.circle(img,(400,50),30,(255,255,0),5)
```

### Add Text

This code adds text to the image img. The text " OPENCV " is placed at coordinates `(300, 200)` (top-left corner of the text baseline). The font used is `cv2.FONT_HERSHEY_COMPLEX` with a scale factor of 1. The text color is a dark green `((0, 150, 0) in BGR format)`, and the text thickness is 3 pixels.

```python
cv2.putText(img, " OPENCV  ", (100, 200), cv2.FONT_HERSHEY_COMPLEX, 2, (0, 150, 0), 3)
plt.imshow(img)
```
